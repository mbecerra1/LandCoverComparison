## Load basemap

```python
import leafmap
import rasterio
```

```python
import folium
import leafmap.foliumap as leafmap_folium
```

## Creating a Split Map for Comparison


```python
# Set up NLCD 2021 Impervious L48 as left layer
# Set up NLCD 2001 Impervious L48 as left layer

m = leafmap_folium.Map(center=[40, -100], zoom=4)

url1 = "https://www.mrlc.gov/geoserver/mrlc_display/NLCD_2021_Land_Cover_L48/wms?"
url2 = "https://www.mrlc.gov/geoserver/mrlc_display/NLCD_2001_Land_Cover_L48/wms?"

left_layer = folium.WmsTileLayer(
    url=url1,
    layers="NLCD_2021_Land_Cover_L48",
    name="NLCD 2021",
    attr="MRLC",
    fmt="image/png",
    transparent=True,
)
right_layer = folium.WmsTileLayer(
    url=url2,
    layers="NLCD_2001_Land_Cover_L48",
    name="NLCD 2001",
    attr="MRLC",
    fmt="image/png",
    transparent=True,
)

left_layer
m.split_map(left_layer, right_layer)
m.add_legend(builtin_legend="NLCD")
m.add_title("NLCD Land Cover Comparison 2021 - 2001", align="center")

m
```

## Export to Html

```python
m.to_html("US_LandCover.html", title="LandCover US Map", width="100%", height="100%")
m
```
